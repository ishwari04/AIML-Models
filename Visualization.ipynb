# Data visualization

```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","March","April","May"]
energy_consp = [1200,1300,1400,1500,1600]

#create aline plot
plt.plot(months,energy_consp,marker='o',color='b',linestyle='--')

plt.title('Energy consumption Over 6 Months')
plt.xlable('Month')
plt.ylable('Energy consumption (MWh)')
plt.show()
```

```python
# bar chart
plt.bar(months,energy_consp,color='b')
plt.title('Energy consumption Over 6 Months')
plt.xlable('Month')
plt.ylable('Energy consumption (MWh)')
plt.show()
```

```python
import matplotlib.pyplot as plt
months = ['Jan','Feb','Mar','Apr','May','Jun']
energy_consumption = [1200,1300,1100,1500,1400,1600]

plt.pie(energy_consumption, labels=months, autopct='%1.1f%%',colors=['gold','lightblue','lightgreen','coral','red','blue'])

plt.title('Energy Consumption Over 6 Months')
plt.show()
```

```python
#scatter plot
carbon_emissions = [400,500,600,300,200] 
plt.scatter(energy_consp,carbon_emissions,color='red')

plt.title('Energy Consumption Vs Carbon_emissions')
plt.show()
```

```python
#histogram
import matplotlib.pyplot as plt
import numpy as np
data = np.random.normal(170,10,250)
plt.hist(data,bins=30)
plt.show();
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    "Solar" : [10,20,30],
    "Wind" : [34,35,40],
    "Hydro" : [50,60,70]

}

df = pd.DataFrame(data)
#create pair plot
sns.pairplot(df)
plt.show()
```

## HeatMap

```python
#we use heatmap to check collinearity
#correlation matrix
cm = df.corr()

sns.heatmap(cm,annot=True,cmap='coolwarm',linewidth=0.5)

plt.show()
```

```python
# create box plot
sns.boxplot(data=df)
plt.show()
```

```python
#violin plot
#plot having one continous feature and categorical 

sns.violinplot(data=df)
plt,show()
```

```python
#reg plot
import seaborn as sns
import matplotlib.pyplot as plt
energy_values = [100,200,300,400,500]
carbon_emissions = [10,20,25,40,50]
df_reg = pd.DataFrame({
    'Energy Consumption': energy_values,
    'Carbon Emissions': carbon_emissions
})
sns.regplot(x='Energy Consumption',y='Carbon Emissions',data=df_reg)
```

## 3d plot

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure()
ax=plt.axes(projection='3d')
z = np.linspace(0,1,1000)
x=z*np.sin(25*z)
y=z*np.cos(25*z)
ax.plot3D(x,y,z,'green')
ax.set_title('3D line plot')
plt.show()
```
